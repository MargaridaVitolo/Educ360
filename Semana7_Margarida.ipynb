**Criação das tabelas**

```python
conn = sqlite3.connect('banco.db')
cursor = conn.cursor()

# Drop tables if they exist
cursor.execute("DROP TABLE IF EXISTS vendas")
cursor.execute("DROP TABLE IF EXISTS produtos")


# Create the 'produtos' table
cursor.execute("""
CREATE TABLE produtos (
    id_produto INTEGER PRIMARY KEY,
    nome_produto TEXT,
    categoria TEXT,
    preco_unitario FLOAT
)
""")

# Create the 'vendas' table
cursor.execute("""
CREATE TABLE vendas (
    id_venda INTEGER PRIMARY KEY,
    id_produto INTEGER,
    quantidade INTEGER,
    data_venda DATE,
    desconto FLOAT,
    FOREIGN KEY (id_produto) REFERENCES produtos(id_produto)
)
""")

conn.commit()
conn.close()
```

##**Etapa 3**

```python
import pandas as pd
```

```python
query = """SELECT v.data_venda,
                  v.quantidade,
                  v.desconto,
                  p.nome_produto,
                  p.categoria,
                  p.preco_unitario,
                  ((COALESCE(v.quantidade,0)*COALESCE(p.preco_unitario,0))-COALESCE(v.desconto,0)) valor_total
           FROM vendas v
           JOIN produtos p ON v.id_produto = p.id_produto
           ORDER BY v.id_venda
      """

df_resultado = pd.read_sql_query(query, conn)
print("\nConsulta com valor_total calculado:")
print(df_resultado.head(10))  # mostrar os 10 primeiros registros
```

##**Etapa 4**

```python
#Tipos dados
print("Tipos de dados:")
print(df_resultado.dtypes)
#Quantidade de registros
print("\nQuantidade de registros:")
print(len(df_resultado))
#Quantidade de registros nulos
print("\nQuantidade de registros nulos:")
print(df_resultado.isnull().sum())
```

```python
#troca do tipo da quantidade para inteiro e colocando o valor como 0 estes campos serão desconsiderados no proximo passo
df_resultado['quantidade'] = df_resultado['quantidade'].fillna(0).astype(int)
#preencher com 0 os campos de desconto que estejam nulo
df_resultado['desconto'] = df_resultado['desconto'].fillna(0)
#colocar a categoria como "Não informada"
df_resultado['categoria'] = df_resultado['categoria'].fillna('Não Informado')
#preencher com 0 os campos de preco_unitario que estejam nulo -  estes campos serão desconsiderados no proximo passo
df_resultado['preco_unitario'] = df_resultado['preco_unitario'].fillna(0)
df_resultado
```

```python
df = df_resultado.query('valor_total > 0')
df
```

##**Etapa 5**

```python
print("\n\033[31mAnálise e Indicadores:\033[0m")
print("=" * 80)
# Total de vendas (quantidade e valor financeiro)
print("\n\033[34mTotal de vendas(qtd):\t\tTotal de Vendas (R$)\033[0m\n")
total_vendas = df['quantidade'].sum()
total_vl_finc = df['valor_total'].sum()
print(f"{total_vendas}\t\t\t\t{total_vl_finc:,.2f}")
print("-" * 80)
# Faturamento médio por categoria
fat_cat = df.groupby('categoria')['valor_total'].mean()
print("\n\033[34mFaturamento médio por categoria:\033[0m\n")
fat_cat_formatted = fat_cat.apply(lambda x: f'{x:,.2f}')
fat_cat_formatted.index.name = None
print(fat_cat_formatted.to_string())
print("-" * 80)

# Produtos mais vendidos
top_prod = df.groupby('nome_produto')['quantidade'].sum().sort_values(ascending=False)
print("\n\033[34mProdutos mais vendidos (Top 5):\033[0m\n")
top_prod_formatted = top_prod
top_prod_formatted.index.name = None
print(top_prod_formatted.head(5).to_string())
print("-" * 80)

# Dias com maior volume de vendas
top_dia = df.groupby('data_venda')['quantidade'].sum().sort_values(ascending=False)
print("\n\033[34mDias com maior volume de vendas (Top 5):\033[0m\n")
print(top_dia.head(5).to_string())
print("-" * 80)

# Categoria mais vendida
top_cat = df.groupby('categoria')['quantidade'].sum().sort_values(ascending=False)
print("\n\033[34mCategoria mais vendida:\033[0m\n")
print(top_cat.to_string())
print("-" * 80)

# Produto com maior desconto
top_desc = df.groupby('nome_produto')['desconto'].sum().sort_values(ascending=False)
print("\n\033[34mProduto com maior desconto (Top 5):\033[0m\n")
print(top_desc.head(5).to_string())
print("=" * 80)
```
