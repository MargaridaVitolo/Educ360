**Conexão e Criação do banco**

```python
conn = sqlite3.connect('banco.db')
try:
  cursor = conn.cursor()
  print("Conexão com o banco de dados bem sucedida!")
except sqlite3.Error as e:
  print(f"Erro ao conectar ao banco de dados: {e}")
```

**Criação das tabelas**

```python
conn = sqlite3.connect('banco.db')
cursor = conn.cursor()

# Drop tables if they exist
cursor.execute("DROP TABLE IF EXISTS vendas")
cursor.execute("DROP TABLE IF EXISTS produtos")


# Create the 'produtos' table
cursor.execute("""
CREATE TABLE produtos (
    id_produto INTEGER PRIMARY KEY,
    nome_produto TEXT,
    categoria TEXT,
    preco_unitario FLOAT
)
""")

# Create the 'vendas' table
cursor.execute("""
CREATE TABLE vendas (
    id_venda INTEGER PRIMARY KEY,
    id_produto INTEGER,
    quantidade INTEGER,
    data_venda DATE,
    desconto FLOAT,
    FOREIGN KEY (id_produto) REFERENCES produtos(id_produto)
)
""")

conn.commit()
conn.close()
```

**Validação Produtos e Vendas**

```python
conn = sqlite3.connect('banco.db')
cursor = conn.cursor()
sel_prod = "select * from produtos"
cursor.execute(sel_prod)
produtos = cursor.fetchall()
sel_vend = "select * from vendas"
cursor.execute(sel_vend)
vendas = cursor.fetchall()
print("-" * 50)
print("PRODUTOS")
print("-" * 50)
for linha in produtos:
    print(linha)
print("-" * 50)
print("VENDAS")
print("-" * 50)
for linha in vendas:
    print(linha)
```
