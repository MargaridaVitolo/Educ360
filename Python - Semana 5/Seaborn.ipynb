```python
import seaborn as sns
import matplotlib.pyplot as plt
# derivado do matplotlib com várias funções já embutidas



```

```python
dados = sns.load_dataset('tips')

print(dados.describe(include= "all"))
```

```python
sns.scatterplot(x="total_bill", y="tip", data=dados)
plt.title("Relação entre Total da Conta e Gorjeta")
plt.show()
```

```python
#regressão
sns.regplot(x="total_bill", y="tip", data=dados)
plt.title("Relação entre Total da Conta e Gorjeta")
plt.show()
```

```python
#barras
sns.barplot(x="day", y="tip", data=dados, estimator = "sum", palette='pastel') #mean
plt.title("Gorjeta por dia da Semana")
plt.show()
```

```python
#boxplot
sns.boxplot(x="day", y="total_bill", data=dados)
plt.title("Distribuição de valor das contas por dia")
plt.show()
```

```python
#histograma
sns.histplot(data=dados, x="total_bill", bins=10, kde=True, color="orange")
#bins equivale a densidade (numero de contas que são apresentadas)
#kde = true fica mais opaco
plt.title("Histograma de valor das contas")
plt.show()
```

```python
#violino
sns.violinplot(x="day", y="tip", data=dados)
plt.title("Distribuição de valor das contas por dia")
plt.show()
```

```python
#linhas
op = 'Female'
dadosf = dados.sex == op
sns.lineplot(x="total_bill", y="tip", data=dados[dadosf], marker="o")
plt.title(f"Relação entre Total da Conta e Gorjeta - Sexo {op}")
plt.show()
```

```python
#hue - similar a um agrupador - não precisa do filtro
sns.lineplot(x="total_bill", y="tip", data=dados, marker="o", hue = 'sex')
plt.title(f"Relação entre Total da Conta e Gorjeta - Sexo {op}")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/alunos_cursos.csv',sep=';')
```

```python
#quantos alunos tenho por curso
sns.countplot(x="Curso", data=df)
plt.title("Alunos por Curso")
#plt.grid(True)
plt.show()
```

```python
#média de notas por curso
sns.lineplot(x="Curso", y="Nota", data=df, marker="o", hue="Certificado")
plt.title("Média de notas por Curso")
plt.grid(True)
plt.show()
```

```python
sns.lineplot(x="Nota", y="Aluno", data=df, marker="o", hue="Curso")
plt.title("Média de notas por Curso")
plt.grid(True)
plt.show()
```

```python
var_curso = 'Excel'
df_curso = df.Curso == var_curso
sns.lineplot(x="Nota", y="Aluno", data=df[df_curso], marker="o")
plt.title("Média de notas por Curso")
plt.grid(True)
plt.show()
```

```python
var_curso = 'Excel'
var_curso2 = 'Python'
df_curso = (df.Curso == var_curso) | (df.Curso == var_curso2)
sns.lineplot(x="Nota", y="Aluno", data=df[df_curso], marker="o", hue="Curso")
plt.title(f"Média de notas por Curso - {var_curso} - {var_curso2}")
plt.grid(True)
plt.show()
```
