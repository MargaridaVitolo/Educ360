# **Import**

```python
import sqlite3
```

**Conexão**

```python
conn = sqlite3.connect('banco.db')
cursor = conn.cursor()
# fazer try catch para validar a conexão
```

**Create Table**

```python
query = "create table inscricoes (nome TEXT NOT NULL, email TEXT NOT NULL, curso TEXT NOT NULL, idade INT)"
cursor.execute(query)

```

**Select**

```python
sel_query = "select * from inscricoes"
cursor.execute(sel_query)
for linha in cursor:
    print(linha)
```

**Insert**

```python
ins_query = "insert into inscricoes values ('Ivan', 'ivan@gmail.com.br','Java', 50)"
#ins_query = "insert into inscricoes (nome, email, curso) values ('Ivan', 'ivan@gmail.com.br','Java')
cursor.execute(ins_query)
# rollback
conn.commit()
```

**Insert via lista**

```python
dados = [
    ('Denise', 'denise@gmail.com','Java',45),
    ('Carlos', 'carlos@gmail.com','Python',35),
    ('Joana', 'joana@gmail.com','Java',24),
    ('Silvio', 'silvio@gmail.com','Dados',67)
]

ins_query = "insert into inscricoes values (?,?,?,?)"
cursor.executemany(ins_query, dados)
conn.commit()
#
```

**Insert via arquivo**

```python
arquivo = "dados.txt"
for linhas in open(arquivo,'r'):
    registro = linhas.split(',')
    print(registro)
    ins_query = "insert into inscricoes values (?,?,?,?)"
    cursor.execute(ins_query, registro)
conn.commit()
```
