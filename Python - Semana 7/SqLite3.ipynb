# **Import**

```python
import sqlite3
```

**Conexão**

```python
conn = sqlite3.connect('banco.db')
cursor = conn.cursor()
# fazer try catch para validar a conexão
```

**Create Table**

```python
query = "create table inscricoes (nome TEXT NOT NULL, email TEXT NOT NULL, curso TEXT NOT NULL, idade INT)"
cursor.execute(query)

```

**Select**

```python
sel_query = "select * from inscricoes"
cursor.execute(sel_query)
for linha in cursor:
    print(linha)
```

**Insert**

```python
ins_query = "insert into inscricoes values ('Ivan', 'ivan@gmail.com.br','Java', 50)"
#ins_query = "insert into inscricoes (nome, email, curso) values ('Ivan', 'ivan@gmail.com.br','Java')
cursor.execute(ins_query)
# rollback
conn.commit()
```

**Modelagem com Join**

```python
query = "create table cursos (idcurso TEXT NOT NULL, nomecurso TEXT NOT NULL, instrutor TEXT NOT NULL)"
cursor.execute(query)
```

```python
dados = [
  ('1','Java','Daniel'),
  ('2','Python', 'Ivan'),
  ('3','Dados','Marco'),
  ('4','Excel','Rogerio'),
  ('5','BI','Carlos')
]

ins_query = "insert into cursos values (?,?,?)"
cursor.executemany(ins_query, dados)
conn.commit()
```

```python
sel_query = "select * from cursos"
cursor.execute(sel_query)
dados = cursor.fetchall()
for linha in dados:
    print(linha)
```

```python
# Alterar o campo curso da tabela inscricoes com base na tabela cursos

upd_query = "update inscricoes set curso = ("
upd_query = upd_query + "select idcurso from cursos c where c.nomecurso = inscricoes.curso)"
upd_query = upd_query + "where exists (select 1 from cursos c where c.nomecurso = inscricoes.curso)"

cursor.execute(upd_query)
conn.commit()
```

```python
query = "select i.nome, i.email, i.idade, c.nomecurso,c.instrutor  from inscricoes i join cursos c on i.curso = c.idcurso"
#query = query + " where c.nomecurso = 'Java' order by i.nome"
cursor.execute(query)
dados = cursor.fetchall()
for linha in dados:
    print(linha)
```
